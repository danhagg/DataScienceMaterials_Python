```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
```

```python
x1 = [1, 2, 3, 4]
y1 = [10, 20, 25, 30]
x2 = [0.3, 3.8, 1.2, 2.5]
y2 = [11, 25, 9, 26]
```

```python
fig = plt.figure()
ax = fig.add_subplot(111)
ax.plot(x1, y1, color='lightblue', linewidth=3)
ax.scatter(x2, y2, color='darkgreen', marker='^')
ax.set_xlim(0, 5)
plt.show()
```

```python
from bokeh.plotting import figure, show, output_notebook
```

```python
from bokeh.models import Range1d
p = figure(plot_height=400, plot_width=400)
output_notebook()

p.line(x1,y1)
p.circle(x2,y2)
p.x_range=Range1d(0,5)
show(p)
```
