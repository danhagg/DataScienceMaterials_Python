```python
import numpy as np
import pandas.io.sql as psql

import psycopg2

#Create a database session
conn = psycopg2.connect(database='stocks', user='danielhaggerty', password='Plikington01')

#Create a client cursor to execute commands
cursor = conn.cursor()

#The variables placeholder must always be a %s, psycop2 will automatically convert the values to SQL literal
query = "SELECT * FROM stocks_table"
psql_data = psql.read_sql(query, conn)
```

```python
# bokeh hist of above
from bokeh.plotting import figure, show, output_notebook
from numpy import histogram, linspace
from scipy.stats.kde import gaussian_kde

output_notebook()

bins = range(0,1010,10)

p = figure(plot_width=600, plot_height=300, title='Histogram of stock Prices, Downloaded from SQL, plotted in bokeh')

p.xaxis.axis_label='Stock Price ($)'
p.yaxis.axis_label='Normalised Count'

# plot actual hist for comparison
hist, edges = histogram(psql_data.AMZN, density=True, bins=bins)
hist1, edges1 = histogram(psql_data.DUK, density=True, bins=bins)
hist2, edges2 = histogram(psql_data.KO, density=True, bins=bins)
p.quad(top=hist, bottom=0, left=edges[:-1], right=edges[1:], alpha=0.4,legend='AMZN')
p.quad(top=hist1, bottom=0, left=edges1[:-1], right=edges1[1:], color='firebrick', alpha=0.4, legend='DUK')
p.quad(top=hist2, bottom=0, left=edges2[:-1], right=edges2[1:], color='green', alpha=0.4, legend='KO')

show(p)
```
