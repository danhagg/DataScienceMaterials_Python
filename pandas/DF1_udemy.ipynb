```python
# Sometimes we wish to change datatype of a col and avoid this...
# ValueError: Cannot convert non-finite values (NA or inf) to integer

# nba.fillna(value=None, method=None, axis=None, inplace=False, limit=None, downcast=None, **kwargs)
nba['Salary'].fillna(0,inplace=True)
```

```python
# Ranks needs all values to be numbers
# Hence the NaN cleaning
nba.dropna(how='all')
nba['Salary'] = nba['Salary'].fillna(0)
nba.head()
```

```python
nba['Salary_Rank'] = nba['Salary'].rank(ascending=False).astype("int")
nba.sort_values(by="Salary_Rank", inplace=True)
nba.head()
```
