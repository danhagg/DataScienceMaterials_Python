```python
sales = pd.read_csv('udemy_material/salesmen.csv', parse_dates=['Date'])
sales.head()
```

```python
sales['Salesman'] = sales['Salesman'].astype('category')
sales.head()
```

```python
sales.dtypes
```

```python
sales_new = sales.pivot(index='Date', columns='Salesman',values='Revenue')
sales_new.head()
```

```python
%matplotlib inline
sales_new.plot(kind='line')
```

```python
sales_new.loc['2016-06-30':'2016-07-30'].plot(kind='line')
```
